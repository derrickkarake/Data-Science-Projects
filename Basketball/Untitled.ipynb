```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import re

Data = pd.read_csv('Datasets/nba_shotchartdetail_2018-19.csv')

Shots_Data = Data[['TEAM_ID','LOC_X','LOC_Y','SHOT_MADE_FLAG']]




Data.head()

```

```python
#Finding the shot zone


```

```python
#Shots_Data.loc[(Shots_Data['LOC_X'] >= 22) | (Shots_Data['LOC_Y'] <= 7.8), "C"] = "C3"





Shots_Data['X'] = Shots_Data.LOC_X / 10
Shots_Data['Y'] = Shots_Data.LOC_Y / 10

Shots_Data["Team_ID"] = '1610612755'


      
```

```python


shots_try = Shots_Data[['TEAM_ID','X','Y','SHOT_MADE_FLAG']]
shots_try.head()
#shots_try

df = pd.DataFrame(shots_try)

shots_try2 = shots_try[:3000]




final_shots = shots_try2.query('TEAM_ID == 1610612755 | TEAM_ID == 1610612738 ')




final_shots['X'] = abs(final_shots['X'])
final_shots['Y'] = abs(final_shots['Y'])

final_shots['Distance'] = ((((final_shots['X'] - 0)**2) + ((final_shots['Y'] - 0)**2)))** 0.5

final_shots




def zones(final_shots):
    if(final_shots['Distance'] >= 23.75 and final_shots['Y'] >= 7):
        return 'NC3'
    elif(final_shots['Distance'] >= 22 and final_shots['Y'] <= 7.8):
        return 'C3'
    else :
        return '2PT'
    
    
    
final_shots['zones'] = final_shots.apply(zones, axis = 1)

    

    
final_shots['zones'] = final_shots.apply(zones, axis = 1)

list_zones = final_shots['zones'].value_counts('percent')
 


    
```

```python
#counting points
def points_made(final_shots):
    if(final_shots['zones'] == '2PT') :
        return '2'
    else :
        return '3'

final_shots['points_made'] = final_shots.apply(points_made, axis = 1)

final_shots.head()


final_shots['points_made'].value_counts()


final_shots.head()

```
